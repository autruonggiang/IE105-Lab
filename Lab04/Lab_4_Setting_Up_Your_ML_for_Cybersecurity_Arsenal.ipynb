Cho biết cảm nghĩ về việc dùng 4 thư viện này

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Read data
X = np.array([[3], [5], [7], [9], [11]]).reshape(-1, 1)
y = [8.0, 9.1, 10.3, 11.4, 12.6]
new_data = np.array([[13]])

# Define model
model = LinearRegression()
model.fit(X, y)

# Predict
pred = model.predict(new_data)
print('Model Prediction for new data: $%.2f' %pred[0] ) #13.73

# Visualize
plt.scatter(X, y)
plt.plot(X, y, color='red')
plt.show()
```

```python
import tensorflow as tf

# Read data
X = np.array([[3], [5], [7], [9], [11]]).astype(np.float32)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6]).astype(np.float32)

# Define model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(1, activation='linear')
])

# Ptimize loss funtion
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Train model
model.fit(X, y, epochs=100)

# Predict
new_data = np.array([[13.0]])
predictions = model.predict(new_data)
print(predictions, "%") #16.436716

# Visualize
plt.scatter(X, y)
plt.plot(X, y, color='red')
plt.show()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Read data
X = np.array([[3], [5], [7], [9], [11]]).astype(np.float32)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6]).astype(np.float32)
new_data = np.array([[13]])

model = Sequential()
model.add(Dense(units=1, input_dim=1))

# Optimize loss fn
optimizer = keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Train model
model.fit(X, y, epochs=100)

# Compile và fit model
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, y, epochs=500)

# Predict
pred = model.predict(new_data)
print(pred) #16.52888

# Visualize
plt.scatter(X, y)
plt.plot(X, y, color='red')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Read data
X = np.array([[3], [5], [7], [9], [11]])
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6])
new_data = np.array([[13]])

X_tensor = torch.from_numpy(X).float()
y_tensor = torch.from_numpy(np.array(y)).float()
X_test = torch.from_numpy(new_data).float()

# Define model and loss fn
model = nn.Linear(in_features=1, out_features=1)
optimizer = optim.SGD(model.parameters(), lr=0.01)
criterion = nn.MSELoss()

# Train model
for epoch in range(100):
    predictions = model(X_tensor)
    loss = criterion(predictions, y_tensor.view(-1, 1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Predict
with torch.no_grad():
    pred = model(X_test).numpy()

print(pred) #16.184908

# Visualize
plt.scatter(X, y)
plt.plot(X, y, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron

# Read data
df = pd.read_csv(root_path + 'sms_spam_perceptron.csv')
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
X = df.iloc[:, [1, 2]].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Define model, train and predict
p = Perceptron(max_iter=40, eta0=0.1, random_state=0)
p.fit(X_train, y_train)
y_pred = p.predict(X_test)

from sklearn.metrics import accuracy_score
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred)) #0.77
```

5. Sinh viên cho biết chức năng của phương thức genfromtxt() trong thư viện
numpy.

```python
from sklearn import *
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Read data
phishing_dataset = np.genfromtxt(root_path + 'phishing_dataset.csv', delimiter=',',dtype=np.int32)
samples = phishing_dataset[:,:-1]
targets = phishing_dataset[:, -1]

# Define model, train and predict
training_samples, testing_samples, training_targets, testing_targets = train_test_split(samples, targets, test_size=0.2, random_state=0)
log_classifier = LogisticRegression()
log_classifier.fit(training_samples, training_targets)
predictions = log_classifier.predict(testing_samples)

accuracy = 100.0 * accuracy_score(testing_targets, predictions)
print ("Logistic Regression accuracy: " + str(accuracy)) #91.67797376752601
```

6. Sinh viên hoàn thiện code Decision Trees trên và đánh giá kết quả nhận được so
với phương pháp Logistic regression.

```python
from sklearn import *
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import tree

# Read data
phishing_dataset = np.genfromtxt(root_path + 'phishing_dataset.csv', delimiter=',',dtype=np.int32)
samples = phishing_dataset[:,:-1]
targets = phishing_dataset[:, -1]
training_samples, testing_samples, training_targets, testing_targets = train_test_split(samples, targets, test_size=0.2, random_state=0)

# Define model, train and predict
tree_classifier = tree.DecisionTreeClassifier()
tree_classifier.fit(training_samples, training_targets)
predictions = tree_classifier.predict(testing_samples)

accuracy = 100.0 * accuracy_score(testing_targets, predictions)
print ("Decision Trees accuracy: " + str(accuracy))#96.42695612844867
```

Dựa vào kết quả nhận được, nhận thấy phương pháp Decision Trees cho độ chính xác cao hơn Logistic regression.
